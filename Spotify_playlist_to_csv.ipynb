```python
#Libraries
from html.parser import HTMLParser
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

# Set-up
name = input("Playlist URL here: ")
names = []
songs = []

#Method
with urllib.request.urlopen(name) as url:
    s = url.read()
soup = BeautifulSoup(s, "html.parser")
soup2 = soup = BeautifulSoup(s, "lxml")

artists_names = soup.find_all("span",class_="Type__StyledComponent-sc-1ell6iv-0 Mesto-sc-1e7huob-0 Row__Subtitle-sc-brbqzp-1 fGDcfx fMlCXO")
for artists in artists_names:
    names.append(artists.text.strip())    
    
song_title = soup.find_all("a",class_="EntityRowV2__AnchorLink-sc-ayafop-8 cGeiev")
for music in song_title:
    songs.append(music.text.strip())
        
# Pandas    
Playlist = pd.DataFrame()
Playlist["Song Title"] = songs
Playlist["Artist"] = names
Playlist.index = Playlist.index + 1
Playlist

#Export Playlist -> .csv
Playlist.to_csv("Playlist.csv",index=False)

#Display list:
Playlist    
```
